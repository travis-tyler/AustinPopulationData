```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '../../Appendix_A_Resources/BEA.gov/CAINC1 Personal Income Summary Personal Income, Population, Per Capita_multi-cities.csv'
path2 = '../../Appendix_A_Resources/BEA.gov/CAINC4 Personal Income and Employment by Major Component_multi-cities.csv'

df = pd.DataFrame(pd.read_csv(path))
df2 = pd.DataFrame(pd.read_csv(path2))
```

```python
austin_df = df2.loc[df2['GeoName'] == 'Austin-Round Rock-Georgetown, TX (Metropolitan Statistical Area)']
seattle_df = df2.loc[df2['GeoName'] == 'Seattle-Tacoma-Bellevue, WA (Metropolitan Statistical Area)']
boston_df = df2.loc[df2['GeoName'] == 'Boston-Cambridge-Newton, MA-NH (Metropolitan Statistical Area)']
madison_df = df2.loc[df2['GeoName'] == 'Madison, WI (Metropolitan Statistical Area)']
```

```python
years = ['2010','2011','2012','2013','2014','2015','2016','2017','2018']

atx_per_cap = austin_df.loc[austin_df['Description'] == 'Per capita personal income (dollars) 4/']
atx_per_cap = atx_per_cap.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
atx_per_cap['Years'] = years

seattle_per_cap = seattle_df.loc[seattle_df['Description'] == 'Per capita personal income (dollars) 4/']
seattle_per_cap = seattle_per_cap.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
seattle_per_cap['Years'] = years

boston_per_cap = boston_df.loc[boston_df['Description'] == 'Per capita personal income (dollars) 4/']
boston_per_cap = boston_per_cap.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
boston_per_cap['Years'] = years

madison_per_cap = madison_df.loc[madison_df['Description'] == 'Per capita personal income (dollars) 4/']
madison_per_cap = madison_per_cap.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
madison_per_cap['Years'] = years

atx_per_cap
```

```python
merge_1 = pd.merge(atx_per_cap,seattle_per_cap,on='Years',how='outer')
merge_2 = pd.merge(merge_1,boston_per_cap,on='Years',how='outer')
per_cap_all = pd.merge(merge_2,madison_per_cap,on='Years',how='outer')

per_cap_all = per_cap_all.rename(columns = {5:"Austin, TX",86:"Seattle, WA",32:"Boston, MA",59:"Madison, WI"})

per_cap_all = per_cap_all.set_index('Years')

per_cap_all
```

```python
per_cap_all.plot(kind='line',title='Metro Areas Income Per Capita (2010-2018)',figsize=(12,8))
plt.xlabel('Year')
plt.ylabel('Income Per Capita')
plt.savefig('../../Appendix_B_Output_Data/MetroAreasPerCapIncome.png',dpi=100)
```

```python
atx_pop = austin_df.loc[austin_df['Description'] == 'Population (persons) 3/']
atx_pop = atx_pop.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
atx_pop['Years'] = years

seattle_pop = seattle_df.loc[seattle_df['Description'] == 'Population (persons) 3/']
seattle_pop = seattle_pop.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
seattle_pop['Years'] = years

boston_pop = boston_df.loc[boston_df['Description'] == 'Population (persons) 3/']
boston_pop = boston_pop.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
boston_pop['Years'] = years

madison_pop = madison_df.loc[madison_df['Description'] == 'Population (persons) 3/']
madison_pop = madison_pop.drop(columns=['GeoFips','LineCode','Description','GeoName']).T
madison_pop['Years'] = years

pop_merge_1 = pd.merge(atx_pop,seattle_pop,on='Years',how='outer')
pop_merge_2 = pd.merge(pop_merge_1,boston_pop,on='Years',how='outer')
pop_all = pd.merge(pop_merge_2,madison_pop,on='Years',how='outer')

pop_all = pop_all.rename(columns = {4:"Austin, TX",85:"Seattle, WA",31:"Boston, MA",58:"Madison, WI"})

pop_all = pop_all.set_index('Years')

pop_all
```

```python
pop_all.plot(kind='line',title='Metro Areas Population (2010-2018)',figsize=(12,8))
plt.xlabel('Year')
plt.ylabel('Population')
plt.ylim(0,5000000)
plt.savefig('../../Appendix_B_Output_Data/MetroAreasPopulation.png',dpi=100)
```
